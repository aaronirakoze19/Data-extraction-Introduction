Full Data Extraction From realistic sales.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import seaborn as sns
import datetime
import random
```

```python
customers = ["Brad", "Alice", "Eve", "Mallory", "Trent", "Peggy", "Victor", "Walter", "Sybil", "Oscar"]
data = []

start_date = datetime.date(2025, 4, 1)
for i in range(61):
    date = start_date + datetime.timedelta(days=i)
    for _ in range(random.randint(3, 6)):
        hours = random.randint(0, 23)
        minutes = random.randint(0, 59)
        seconds = random.randint(0, 59)
        last_updated = datetime.datetime.combine(date, datetime.time(hours, minutes, seconds))
        data.append({
            "id": random.randint(1000, 9999),
            "customer": random.choice(customers),
            "date": date.isoformat(),
            "amount": round(random.uniform(100, 2000), 2),
            "Last_updated": last_updated.isoformat()
        })

df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)
df.head()
```

```python
df_full = pd.read_csv("sales_data.csv", parse_dates=["Last_updated"])
print("full data construction")
df_full.head()
```

Incremental data extraction simulate last date.

```python
last_extracted_date = datetime.datetime(2025,5, 15)
df_incremental = df_full[df_full["Last_updated"] > last_extracted_date]
print("\nincremental data construction(after {last_extracted_date}):")
df_incremental.head()
```
